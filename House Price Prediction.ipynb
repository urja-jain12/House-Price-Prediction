```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import sklearn.datasets
from sklearn.model_selection import train_test_split
from xgboost import XGBRegressor
from sklearn import metrics
```

Importing The California House Price Datasets

```python
from sklearn.datasets import fetch_california_housing
housing = fetch_california_housing()
```

```python
print(housing)
```

```python
#Loading the dataset to pandas dataframe

house_price_df=pd.DataFrame(housing.data,columns=housing.feature_names)
```

```python
house_price_df.head(5)
```

```python
house_price_df.shape
```

```python
house_price_df['Price'] = housing.target
```

```python
house_price_df.head(5)
```

```python
house_price_df.shape
```

```python
#check for missing values
house_price_df.isnull().sum()
```

```python
#statistical measures of data
house_price_df.describe()
```

Understanding the correlation between various features in the dataset


```python
correlation=house_price_df.corr()
```

```python
#construsting heatmap to understand the correlation
plt.figure(figsize=(10,10))
sns.heatmap(correlation,cbar=True,square=True,fmt='.1f',annot=True,annot_kws={'size':8},cmap='Blues')
```

splitting the data and the target


```python
X=house_price_df.drop(['Price'],axis=1)
Y=house_price_df['Price']
```

```python
house_price_df.columns
```

```python
print(X)
print(Y)
```

Splitting the data into training data and tesing data


```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,random_state=2)
```

```python
print(X.shape,X_train.shape,X_test.shape)
```

Model Training

```python
#XGBoost regressor-->loading the model
model=XGBRegressor()
```

```python
#training the model
model.fit(X_train,Y_train)
```

#Evaluation metric (in prev we used accuracy as an evaluation metric but here accuracy score is not used as it is a regression problem)
# so in regression problems we use the metrics such as mean absolute error and r2 error and all

Prediction on training data

```python
training_data_prediction=model.predict(X_train)
```

```python
print(training_data_prediction)
```

```python
#R square error
score1=metrics.r2_score(Y_train,training_data_prediction)

#Mean Absolute error
score_2=metrics.mean_absolute_error(Y_train,training_data_prediction)

print("R squared error as: ",score1)
print("Mean Absolute Error: ",score_2)
```

Visualising the Actual prices and the predicted prices

```python
plt.scatter(Y_train,training_data_prediction)
plt.xlabel("Actual Prices")
plt.ylabel("Predicted Prices")
plt.title("Actual Price Vs Predicted Price")
plt.show()
```

```python
#prediction on test data
testing_data_prediction=model.predict(X_test)
```

```python
print(testing_data_prediction)
```

```python
#R square error
score1=metrics.r2_score(Y_test,testing_data_prediction)

#Mean Absolute error
score_2=metrics.mean_absolute_error(Y_test,testing_data_prediction)

print("R squared error as: ",score1)
print("Mean Absolute Error: ",score_2)
```
